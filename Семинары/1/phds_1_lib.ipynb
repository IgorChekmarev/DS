### Задача 1 (<a  target="_blank" href="https://numpy.org/doc/stable/">numpy</a>)

```python
import numpy as np
```

**1.**
* Создайте массив чисел от $-2\pi$  до $2\pi$.

* Посчитайте сумму поэлементных квадратов синуса и косинуса для данного массива.

* С помощью `np.allclose` проверьте, что в ответе только единицы.

```python
massive = np.linspace(-2*np.pi, 2*np.pi, 10)

massive_2 = np.zeros(10)
for i in range(len(massive)):
    massive_2[i] = (np.cos(massive[i]))**2 + (np.sin(massive[i]))**2
    
np.allclose(massive_2, 1)
```

**2.** Cоздайте матрицы $\begin{pmatrix} -3 & 4 \\ 4 & 3 \end{pmatrix}$ и $\begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix}$. Посчитайте их поэлементное и матричное произведения.

```python
mat_a = np.array([[-3, 4], [4, 3]])
mat_b = np.array([[2,1], [1,2]])
print(mat_a*mat_b)
print(mat_a@mat_b)
```

**3.** В статистике и машинном обучении часто приходится иметь с функцией $RSS$ (*Residual Sum of Squares*), которая вычисляется по формуле
$$
    RSS = \sum_{i=1}^{n} (y_i - a_i)^2,
$$
где $y_i$ — координаты одномерного вектора $y$,  $a_i$ — координаты одномерного вектора $a$. Посчитайте $RSS$ для $y = (1, 2, 3, 4, 5),\ a = (3, 2, 1, 0, -1)$.

```python
y = (1, 2, 3, 4, 5)
a = (3, 2, 1, 0, -1)

sum = 0
for i in range(len(y)):
    sum += (y[i] - a[i])**2
print(sum)
```

### Задача 3 (<a  target="_blank" href="http://docs.scipy.org/doc/scipy/reference/stats.html">scipy.stats</a>)

```python
import scipy.stats as sps
```

Модуль `scipy.stats` содержит в себе инструменты для работы со многими семествами распределений.

Рассмотрим ${U}(a, b)$ — равномерное распределение

**Основные свойства:**
* математическое ожидание: $\frac{a+b}{2}$,
* дисперсия: $\frac{(b-a)^2}{12}$.
  
**Параметры** в `scipy.stats.uniform`:  
* `loc` = $a$,
* `scale` = $b-a$.

Некоторые методы, которые могут понадобиться:
`.pdf()`,  `.rvs()`.

**1.** Постройте график плотности *стандартного* равномерного распределения ${U}(0, 1)$.

```python
from scipy.stats import uniform
x = np.linspace(-0.5, 1.5, 1000)
pdf_val = uniform.pdf(x, loc = 0, scale = 1)
plt.plot(x, pdf_val)
```

**2.** Сгенерируйте данные из *стандартного* равномерного распределения ${U}(0, 1)$ и сравните гистограмму с плотностью.

```python
from scipy.stats import uniform
x = np.linspace(-0.5, 1.5, 1000)

rvs_val = uniform(loc = 0, scale = 1).rvs(1000)
pdf_val = uniform.pdf(x, loc = 0, scale = 1)
plt.plot(x, pdf_val)
plt.hist(rvs_val, density = True)
```

**3.** Постройте плотности распределений ${U}(0, 1)$, ${U}(0, 2)$, ${U}(4, 1)$ **на одном графике**, опишите как плотность распределения зависит от его параметров.

поменяны границы с U(4, 1) на U(1, 4)

```python
from scipy.stats import uniform
x = np.linspace(-0.5, 4.5, 1000)
pdf_val = uniform.pdf(x, loc = 0, scale = 1)
pdf_val1 = uniform.pdf(x, loc = 0, scale = 2)
pdf_val2 = uniform.pdf(x, loc = 1, scale = 3)
plt.plot(x, pdf_val)
plt.plot(x, pdf_val1)
plt.plot(x, pdf_val2)
```

**Ответ:** чем больше scale, тем меньше значение плотности распределения
