### –ó–∞–¥–∞—á–∞ 2. ü§ñ

–û–±—É—á–∏–º –º–æ–¥–µ–ª—å **–ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏** —Å –ø–æ–º–æ—â—å—é –∫–ª–∞—Å—Å–∞ <a target="_blank" href="https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html">`LogisticRegression`</a>.

#### –í–∞–∂–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–∞

* **`fit_intercept`** ‚Äî –¥–æ–±–∞–≤–ª—è—Ç—å –ª–∏ —Å–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω (–∏–Ω—Ç–µ—Ä—Å–µ–ø—Ç). –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é: `True`.
* **`solver`** ‚Äî –∞–ª–≥–æ—Ä–∏—Ç–º –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é: **`"lbfgs"`** (—ç—Ç–æ –±—ã—Å—Ç—Ä—ã–π —Å–ø–æ—Å–æ–± –Ω–∞—Ö–æ–¥–∏—Ç—å –º–∏–Ω–∏–º—É–º —Ñ—É–Ω–∫—Ü–∏–∏ –±–µ–∑ –ø–µ—Ä–µ–±–æ—Ä–∞. –ù–µ —Ö—Ä–∞–Ω–∏—Ç –æ–≥—Ä–æ–º–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –≤—Ç–æ—Ä—ã—Ö –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã—Ö (–ì–µ—Å—Å–∏–∞–Ω).  –î—Ä—É–≥–∏–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã —Å–º. –≤ –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏.
* **`max_iter`** ‚Äî –º–∞–∫—Å–∏–º—É–º –∏—Ç–µ—Ä–∞—Ü–∏–π –æ–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä–∞. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é: `100`.
* **`penalty`**, **`C`** ‚Äî –ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏. –ü–æ–¥—Ä–æ–±–Ω–æ —Ä–∞–∑–±–µ—Ä—ë–º –≤ –¥—Ä—É–≥–æ–π —Ç–µ–º–µ.

#### –û—Å–Ω–æ–≤–Ω—ã–µ –º–µ—Ç–æ–¥—ã

+ **`fit(X, y)`**

–û—Ü–µ–Ω–∏–≤–∞–µ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–¥–µ–ª–∏ $(\theta)$, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–Ω–æ –æ–±—ä—è—Å–Ω—è—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ, —á—Ç–æ —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å

+ **`predict_proba(X)`**

–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ—Ü–µ–Ω–∫–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∫–ª–∞—Å—Å–æ–≤ $(P(y=0\mid x),; P(y=1\mid x))$. **–í –∑–∞–¥–∞—á–µ –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –¥–ª—è –∫–ª–∞—Å—Å–∞ 1**:
$$
\widehat p(x)=\sigma\big(x^\top\widehat\theta\big)=\frac{1}{1+\exp(-x^\top\widehat\theta)}.
$$
–î–ª—è –∫–∞–∂–¥–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –∏–∑ `X` –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –¥–≤–∞ —á–∏—Å–ª–∞ &mdash; –æ—Ü–µ–Ω–∫–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∫–ª–∞—Å—Å–æ–≤ 0 –∏ 1. –í–∞–∂–Ω–æ –ø–æ–º–Ω–∏—Ç—å, —á—Ç–æ –æ–Ω–∏ –Ω–µ —è–≤–ª—è—é—Ç—Å—è –∏—Å—Ç–∏–Ω–Ω—ã–º–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º–∏, —ç—Ç–æ —Ç–æ–ª—å–∫–æ *–æ—Ü–µ–Ω–∫–∏* –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π. –û–±–æ–±—â–µ–Ω–∏–µ –Ω–∞ –º–Ω–æ–≥–æ–∫–ª–∞—Å—Å–æ–≤—ã–π —Å–ª—É—á–∞–π —Å–º. [—Ç—É—Ç](https://www.google.com/url?q=https://scikit-learn.org/stable/modules/linear_model.html%23multinomial-case&sa=D&source=editors&ust=1759322416262384&usg=AOvVaw2y_su6rexEAVSyAUouWVUK)

+ **`predict(X)`**

–ú–µ—Ç–æ–¥ `predict(X)` –ø–µ—Ä–µ–≤–æ–¥–∏—Ç —ç—Ç–∏ –æ—Ü–µ–Ω–∫–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π, –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –º–µ—Ç–æ–¥–æ–º `predict_proba(X)`, –≤ –±–∏–Ω–∞—Ä–Ω—ã–µ –∫–ª–∞—Å—Å—ã (¬´0¬ª –∏–ª–∏ ¬´1¬ª) —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –ø–æ—Ä–æ–≥–∞, –æ–±—ã—á–Ω–æ —Ä–∞–≤–Ω–æ–≥–æ 0.5. –¢–æ –µ—Å—Ç—å:

$$
\widehat{y}(x) =
\begin{cases}
1, & \text{–µ—Å–ª–∏ } \widehat{p}(x) \geq 0.5 \\
0, & \text{–µ—Å–ª–∏ } \widehat{p}(x) < 0.5
\end{cases}
$$

–≠—Ç–∏ —à–∞–≥–∏ –ø–æ–∑–≤–æ–ª—è—é—Ç –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º—É –∫–ª–∞—Å—Å—É –∏ –ø—Ä–∏–Ω–∏–º–∞—Ç—å —Ä–µ—à–µ–Ω–∏–µ –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –Ω–∞ —ç—Ç–æ–º –æ—Å–Ω–æ–≤–∞–Ω–∏–∏.

  

–° –ø–æ–¥—Ä–æ–±–Ω–æ–π –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–µ–π –∏ –º–µ—Ç–æ–¥–∞–º–∏ –º–æ–∂–Ω–æ –æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è [—Ç—É—Ç](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)


**1.** –°–Ω–∞—á–∞–ª–∞ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ —Å–∞–º—É –º–æ–¥–µ–ª—å, —Ñ–∏–∫—Å–∏—Ä—É–π—Ç–µ `random_state` –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤, –∞ —Ç–∞–∫ –∂–µ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ –ø–∞—Ä–∞–º–µ—Ç—Ä `penalty=None`

```python
logreg = LogisticRegression(random_state=0, penalty=None)
```

–¢–µ–ø–µ—Ä—å –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å

```python
logreg.fit(X_train, y_train)
```

–ò, –Ω–∞–∫–æ–Ω–µ—Ü, —Å–¥–µ–ª–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ, –∞ —Ç–∞–∫ –∂–µ –æ—Ü–µ–Ω–∏–º –∫–∞—á–µ—Å—Ç–≤–æ –Ω–∞—à–µ–π –º–æ–¥–µ–ª–∏ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç—Ä–∏–∫–∏ accuracy –¥–æ 3 –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

```python
y_pred_log_train = logreg.predict(X_train)
y_pred_log_test  = logreg.predict(X_test)

acc_train = accuracy_score(y_train, y_pred_log_train)  
acc_test = accuracy_score(y_test, y_pred_log_test)

print(np.round(acc_train, 3))
print(np.round(acc_test, 3))
```

### –ó–∞–¥–∞—á–∞ 3. ‚è∞

 –ò–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è –º—ã –ø–æ–Ω—è–ª–∏, —á—Ç–æ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –Ω–µ–∫–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ –≤ –º–æ–¥–µ–ª–∏ &mdash; `threshold`, –ø–æ–º–æ–≥–∞—é—â–∞—è –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–∞–≤—Ç—å "–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏", –∫–∞–∫ –ø–æ–ª–Ω–æ—Ü–µ–Ω–Ω—ã–µ –∫–ª–∞—Å—Å—ã  

–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –∏–º–µ–Ω–Ω–æ –æ–Ω —ç—Ç–æ –¥–µ–ª–∞–µ—Ç, –∏ –∫–∞–∫ —ç—Ç–æ –º–æ–∂–µ—Ç –ø–æ–≤–ª–∏—è—Ç—å –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏

**1.** –í–æ–∑—å–º–∏—Ç–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å—É 1, –∑–∞–¥–∞–π—Ç–µ —Å–µ—Ç–∫—É –∑–Ω–∞—á–µ–Ω–∏–π threshold –∏ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ accuracy –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π threshold

```python
def plot_accuracy_threshold(prob_class_1: np.ndarray) -> None:

  """
    –°—Ç—Ä–æ–∏—Ç –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å accuracy –æ—Ç –ø–æ—Ä–æ–≥–∞ –æ—Ç—Å–µ—á–µ–Ω–∏—è –¥–ª—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∫–ª–∞—Å—Å–∞ 1.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
    ----------
    prob_class_1 : np.ndarraye of shape (n_samples,)
        –û—Ü–µ–Ω—ë–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ –∫–ª–∞—Å—Å—É 1.
        –î–æ–ª–∂–Ω—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å —Ç–µ–º –∂–µ –æ–±—ä–µ–∫—Ç–∞–º, —á—Ç–æ –∏ –≥–ª–æ–±–∞–ª—å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è `y_test`.

  """
  thresholds = np.linspace(0, 1, 1000)
  accur = []
  for i in thresholds:
      y_preds = prob_class_1>i

      accur.append(accuracy_score(y_test, y_preds))


  plt.figure(figsize=(18, 6))

  plt.plot(thresholds, accur, label='Standard Model', color='blue')
  plt.title('Accuracy Comparison (Clipped)')
  plt.xlabel('Threshold')
  plt.ylabel('Accuracy')
  plt.ylim(0.5, 1)

  plt.legend()
  plt.grid(True)

  plt.tight_layout()
  plt.show()
```

–û—Ç–æ–±—Ä–∞–∑–∏—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å accuracy –æ—Ç threshold –¥–ª—è "–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π" –ø–µ—Ä–≤–æ–≥–æ –∫–ª–∞—Å—Å–∞

```python
plot_accuracy_threshold(data['Probability_class_1'])
```

- –ú–æ–∂–Ω–æ –ª–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Å–∫–∞–∑–∞—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ö–æ—Ä–æ—à–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –º–æ–¥–µ–ª—å?

**–í—ã–≤–æ–¥:**  

–î–∞, –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –º–µ–∂–¥—É –∫–ª–∞—Å—Å–∞–º–∏ –µ—Å—Ç—å, –ø—Ä–∏—á–µ–º –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã–µ.
–ù–∞–∏–±–æ–ª–µ–µ —Å–ª–æ–∂–Ω—ã–µ –¥–ª—è –º–æ–¥–µ–ª–∏ –æ–±—ä–µ–∫—Ç—ã - —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –∫–ª–∞—Å—Å–∞ 1 –æ–∫–æ–ª–æ 0.5
–î–∞, –ø–æ –≥—Ä–∞—Ñ–∏–∫—É –º–æ–∂–Ω–æ —É–≤–∏–¥–µ—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ö–æ—Ä–æ—à–∞ –º–æ–¥–µ–ª—å. –ï—Å–ª–∏ –±–æ–ª—å—à–∞—è —á–∞—Å—Ç—å –æ–±—ä–µ–∫—Ç–æ–≤ –∏–º–µ–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ –∫–ª–∞—Å—Å—É 1 –æ–∫–æ–ª–æ 1 –∏–ª–∏ 0, —Ç–æ –º–æ–¥–µ–ª—å –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–∞—è. –ß–µ–º –±–æ–ª—å—à–µ –æ–±—ä–µ–∫—Ç–æ–≤ –∏–º–µ—é—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ –∫–ª–∞—Å—Å—É 1 –≤ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ –º–µ–∂–¥—É 0 –∏ 1, –∏ —á–µ–º –±–æ–ª—å—à–µ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è –∫–ª–∞—Å—Å—ã - —Ç–µ–º —Ö—É–∂–µ —Ä–∞–±–æ—Ç–∞–µ—Ç –º–æ–¥–µ–ª—å

–î–ª—è –±–æ—Ä—å–±—ã —Å –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º –µ—Å—Ç—å —Ä–∞–∑–ª–∏—á–Ω—ã–µ –º–µ—Ç–æ–¥—ã, —Ç–∞–∫–∏–µ –∫–∞–∫ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –º–æ–¥–µ–ª–∏ [`class_weight=balanced`](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#:~:text=to%20be%20increased.-,class_weight,-dict%20or%20%E2%80%98balanced) –∏–ª–∏ `down-sampling`. –¢–∞–∫ –∂–µ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å [`stratify`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html#:~:text=must%20be%20None.-,stratify,-array%2Dlike%2C%20default) –ø—Ä–∏ —Ä–∞–∑–±–∏–µ–Ω–∏–∏ —Å –ø–æ–º–æ—â—å—é `train_test_split`, –Ω–æ –æ–± —ç—Ç–æ–º –ø–æ–∑–∂–µ  

–¢—É—Ç –º—ã —Å –≤–∞–º–∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ `down-sampling`, –∞ –≤ –¥–æ–º–∞—à–Ω–µ–º –∑–∞–¥–∞–Ω–∏–∏ –Ω–∞ `class_weight=balanced`

**4.** –ò–¥–µ—è `down-sampling` –ø—Ä–æ—Å—Ç–∞: –ø—Ä–æ—Å—Ç–æ –æ–±—Ä–µ–∂–µ–º –º–∞–∂–æ—Ä–Ω—ã–π (—Ç–æ—Ç –∫–æ—Ç–æ—Ä–æ–≥–æ –±–æ–ª—å—à–µ) –∫–ª–∞—Å—Å –≤ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –¥–æ —Ä–∞–∑–º–µ—Ä–∞ –º–∏–Ω–æ—Ä–Ω–æ–≥–æ

```python
y = np.asarray(y_train).ravel()

# –∏–Ω–¥–µ–∫—Å—ã –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞
idx0 = np.where(y == 0)[0]
idx1 = np.where(y == 1)[0]

# —Ä–∞–∑–º–µ—Ä –º–µ–Ω—å—à–µ–≥–æ –∫–ª–∞—Å—Å–∞
n_min = min(len(idx0), len(idx1))

rng = np.random.default_rng(42)
idx0_ds = rng.choice(idx0, size=n_min, replace=False)
idx1_ds = rng.choice(idx1, size=n_min, replace=False)

# –∏—Ç–æ–≥–æ–≤—ã–µ –∏–Ω–¥–µ–∫—Å—ã (–ø–µ—Ä–µ–º–µ—à–∞–µ–º)
idx_ds = np.concatenate([idx0_ds, idx1_ds])
rng.shuffle(idx_ds)

# downsampled train
X_train_bal = X_train.iloc[idx_ds]
y_train_bal = y[idx_ds]
```

–ü–æ—Å–º–æ—Ç—Ä–∏—Ç–µ –Ω–∞ –Ω–æ–≤—ã–µ –ø–æ–ª—É—á–∏–≤—à–∏–µ—Å—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –≤—ã–±–æ—Ä–æ–∫

```python
sns.kdeplot(X_train_bal)
plt.show()

sns.kdeplot(y_train_bal)
plt.show()
```

–ó–∞–Ω–æ–≤–æ –æ–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª—å –∏ –ø–æ—Å–º–æ—Ç—Ä–∏—Ç–µ —Ç–µ–ø–µ—Ä—å –Ω–∞ –∫–∞—á–µ—Ç—Å–≤–æ `accuracy`, —Ç–∞–∫ –∂–µ –æ–∫—Ä—É–≥–ª–∏–≤ –¥–æ 3 –∑–Ω–∞–∫–∞ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

```python
logreg.fit(X_train_bal, y_train_bal)
y_pred_log_train_bal = logreg.predict(X_train_bal)
y_pred_log_test  = logreg.predict(X_test)

acc_train = accuracy_score(y_train_bal, y_pred_log_train_bal)  
acc_test = accuracy_score(y_test, y_pred_log_test)

print("train accuracy score:", np.round(acc_train, 3))
print("test accuracy score:", np.round(acc_test, 3))
```
